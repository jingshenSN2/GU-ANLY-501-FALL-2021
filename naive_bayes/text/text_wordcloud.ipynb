```python
import pandas as pd

dat = pd.read_csv('text_train_labelled.csv')
dat['LABEL'].value_counts()
```

```python
neutral = ' '.join(dat[dat['LABEL'] == 0]['text'])
neg = ' '.join(dat[dat['LABEL'] == 1]['text'])
pos = ' '.join(dat[dat['LABEL'] == 1]['text'])
```

```python
from matplotlib import pyplot as plt
from wordcloud import WordCloud

plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
wc = WordCloud(background_color="white", height=800, width=800).generate_from_text(neg)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title('Negative')

plt.subplot(1, 3, 2)
wc = WordCloud(background_color="white", height=800, width=800).generate_from_text(neutral)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title('Neutral')

plt.subplot(1, 3, 3)
wc = WordCloud(background_color="white", height=800, width=800).generate_from_text(pos)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title('Positive')
plt.savefig(f'wordcloud.png', dpi=300, bbox_inches='tight')
```
